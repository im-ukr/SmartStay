```python
import datetime
import holidays

# Function to check if a given date is a weekend
def is_weekend(date_str):
    date_obj = datetime.datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S')
    return date_obj.weekday() >= 5  # 5 represents Saturday, 6 represents Sunday

# Function to check if a given date is a public holiday in India
def is_public_holiday(date_str, country='IN'):
    india_holidays = holidays.CountryHoliday(country)
    date_obj = datetime.datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S').date()
    return date_obj in india_holidays

# Example usage
date_to_check = '2024-07-20 05:33:05'

if is_weekend(date_to_check):
    print(f"{date_to_check} was a weekend.")
else:
    print(f"{date_to_check} was not a weekend.")

if is_public_holiday(date_to_check):
    print(f"{date_to_check} was a public holiday in India.")
else:
    print(f"{date_to_check} was not a public holiday in India.")
```

```python
from datetime import datetime
import holidays

def is_weekend(date_str):
    date_obj = datetime.strptime(date_str, '%Y-%m-%d')
    return date_obj.weekday() >= 5  # 5 represents Saturday, 6 represents Sunday

def get_india_public_holidays(year):
    india_holidays = holidays.India(years=year)
    holidays_list = india_holidays.items()
    return holidays_list

def check_weekend_and_holidays(date_str):
    try:
        # Check if the date is a weekend
        is_weekend_flag = is_weekend(date_str)
        
        # Check if the date is a public holiday in India
        date_obj = datetime.strptime(date_str, '%Y-%m-%d')
        year = date_obj.year
        india_holidays = get_india_public_holidays(year)
        is_holiday_flag = date_obj.strftime('%Y-%m-%d') in [date[0].strftime('%Y-%m-%d') for date in india_holidays]
        
        return is_weekend_flag, is_holiday_flag
    
    except Exception as e:
        print(f"Error occurred: {e}")
        return None, None

# Example usage:
date_to_check = '2024-12-25'  # Date in %Y-%m-%d format
is_weekend_flag, is_holiday_flag = check_weekend_and_holidays(date_to_check)

if is_weekend_flag is not None and is_holiday_flag is not None:
    if is_weekend_flag:
        print(f"{date_to_check} falls on a weekend.")
    else:
        print(f"{date_to_check} does not fall on a weekend.")
    
    if is_holiday_flag:
        print(f"{date_to_check} is a public holiday in India.")
    else:
        print(f"{date_to_check} is not a public holiday in India.")
else:
    print("Unable to determine weekend or holiday status.")
```

```python
from datetime import datetime
import holidays

def convert_to_date_format(date_str):
    try:
        datetime.strptime(date_str, '%Y-%m-%d')
        return date_str  # Already in %Y-%m-%d format
    except ValueError:
        try:
            date_obj = datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S')
            return date_obj.strftime('%Y-%m-%d')  # Convert to %Y-%m-%d format
        except ValueError as e:
            print(f"Error: {e}. Invalid date format provided.")
            return None

def is_weekend(date_str):
    date_obj = datetime.strptime(date_str, '%Y-%m-%d')
    return date_obj.weekday() >= 5  # 5 represents Saturday, 6 represents Sunday

def get_india_public_holidays(year):
    india_holidays = holidays.India(years=year)
    holidays_list = india_holidays.items()
    return holidays_list

def check_weekend_and_holidays(date_str):
    try:
        # Convert date_str to %Y-%m-%d format if needed
        date_str = convert_to_date_format(date_str)
        if date_str is None:
            return None, None
        
        # Check if the date is a weekend
        is_weekend_flag = is_weekend(date_str)
        
        # Check if the date is a public holiday in India
        date_obj = datetime.strptime(date_str, '%Y-%m-%d')
        year = date_obj.year
        india_holidays = get_india_public_holidays(year)
        is_holiday_flag = date_obj.strftime('%Y-%m-%d') in [date[0].strftime('%Y-%m-%d') for date in india_holidays]
        
        return is_weekend_flag, is_holiday_flag
    
    except Exception as e:
        print(f"Error occurred: {e}")
        return None, None

# Example usage:
date_to_check = '2024-03-25 05:33:05'  # Date in %Y-%m-%d %H:%M:%S format
is_weekend_flag, is_holiday_flag = check_weekend_and_holidays(date_to_check)

if is_weekend_flag is not None and is_holiday_flag is not None:
    if is_weekend_flag:
        print(f"{date_to_check} falls on a weekend.")
    else:
        print(f"{date_to_check} does not fall on a weekend.")
    
    if is_holiday_flag:
        print(f"{date_to_check} is a public holiday in India.")
    else:
        print(f"{date_to_check} is not a public holiday in India.")
else:
    print("Unable to determine weekend or holiday status.")
```
